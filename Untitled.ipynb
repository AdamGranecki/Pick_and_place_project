```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Wczytaj buffer
data = np.load("buffer.npz", allow_pickle=True)
imgs, goals = data["imgs"], data["goals"]

# Sprawdź kształty
imgs.shape, goals.shape
```

```python
# Wyświetlimy wszystkie 5 próbek z bufora: obraz RGB + odpowiadające współrzędne

fig, axes = plt.subplots(5, 2, figsize=(8, 15))

for i in range(5):
    rgb = imgs[i, :, :, :3]  # RGB
    depth = imgs[i, :, :, 3]                # kanał głębokości
    
    # RGB
    axes[i,0].imshow(rgb)
    axes[i,0].axis("off")
    axes[i,0].set_title(f"Przykład {i+1} - RGB")
    
    # Współrzędne
    axes[i,1].imshow(depth, cmap="gray", vmin=0.0, vmax=1.0)
    axes[i,1].axis("off")
    axes[i,1].set_title(f"Goal: {goals[i]}")

    print("Maksymalna wartość w depth[0]:", np.max(depth))
    print("Minimalna wartość w depth[0]:", np.min(depth))

plt.tight_layout()
plt.show()

```
